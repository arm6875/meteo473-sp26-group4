# New Download (2mTemp, Dewpoint, Cloud Cover, CAPE & CIN)

```python
from herbie import Herbie, FastHerbie
import pandas as pd, numpy as np
import xarray as xr
import dask
```

```python
#Date range for FastHerbie
DATES = pd.date_range(
    start="2026-04-15 00:00",
    periods=41,
    freq="1h",
)

# Create a range of forecast lead times
fxx = list(range(0, 24))

DATES, fxx
```

```python
FH = FastHerbie(DATES, model="hrrr", fxx=fxx)
FH
```

```python
FH.objects, len(FH.objects)
```

```python
run = "2026-04-15 00:00" #must be 00:00 to not drop files.
```

```python
H = FastHerbie([run], model="hrrr", fxx=np.arange(0,49,1).tolist(), save_dir='meteo473/sp26_groupwork/473_sp26_group4/NewVariableDownloads', overwrite=True)
```

```python
H.inventory(':DPT:2 m above')
```

```python
H.inventory(':CIN:surface:')
```

```python
H.inventory(':UGRD:10 m above')
```

```python
H.inventory(':VGRD:10 m above')
```

```python
ss = "(:TMP:2 m|:CAPE:surface:|:TCDC:entire|:CIN:surface:|:UGRD:10 m above|:VGRD:10 m above|:DPT:2 m above)"
H.inventory(ss)
```

```python
fp = H.download(ss)
```

```python
fp
```

```python
ds = xr.open_mfdataset(fp, combine='nested', concat_dim='valid_time')
```

```python
ds
```

```python
ds = ds.sortby('valid_time')
ds
```

```python
filename = 'nowhrrr.nc'
```

```python
ds.to_netcdf(filename)
```
