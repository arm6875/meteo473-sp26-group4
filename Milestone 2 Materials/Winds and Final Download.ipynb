```python
from herbie import Herbie, FastHerbie
import pandas as pd, numpy as np
import xarray as xr
import dask
```

```python
run = "2026-04-15 00:00" #must be 00:00 to not drop files.
```

```python
H = FastHerbie([run], model="hrrr", fxx=np.arange(0,49,1).tolist(), save_dir='meteo473/sp26_groupwork/473_sp26_group4/NewVariableDownloads', overwrite=True)
```

```python
H.inventory(':UGRD:10 m above')
```

```python
H.inventory(':VGRD:10 m above')
```

```python
ss = "(:UGRD:10 m above|:VGRD:10 m above)"
H.inventory(ss)
```

```python
fp = H.download(ss)
```

```python
fp
```

```python
ds = xr.open_mfdataset(fp, combine='nested', concat_dim='valid_time')
```

```python
ds
```

```python
ds = ds.sortby('valid_time')
ds
```

```python
filename = 'u_and_v_nowhrrr.nc'
```

```python
ds.to_netcdf(filename)
```
