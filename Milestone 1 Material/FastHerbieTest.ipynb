```python
from herbie import Herbie, FastHerbie
import pandas as pd, numpy as np
import xarray as xr
import dask
```

# Fast Herbie

```python
#Date range for FastHerbie
DATES = pd.date_range(
    start="2021-10-07 00:00",
    periods=41,
    freq="6h",
)

# Create a range of forecast lead times
fxx = list(range(0, 24))

DATES, fxx
```

```python
FH = FastHerbie(DATES, model="hrrr", fxx=fxx)
FH
```

```python
FH.objects, len(FH.objects)
```

```python
run = "2021-10-13 00:00" #must be 00:00 to not drop files.
```

```python
H = FastHerbie([run], model="hrrr", fxx=np.arange(0,49,1).tolist(), save_dir='./data/.', overwrite=True)
```

```python
H.inventory('PRATE')
```

```python
ss2 = "(:TMP:2 m|:RH:2 m|:TCDC:entire|LTNG|PRATE)"
H.inventory(ss2)
```

```python
fp = H.download(ss2)
```

```python
fp
```

```python
ds = xr.open_mfdataset(fp, combine='nested', concat_dim='valid_time')
```

```python
ds
```

```python
filename = 'nullhrrr.nc'
```

```python
ds.to_netcdf(filename)
```
